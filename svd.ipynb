# another example of svd

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/drive/MyDrive/Google colab/linear aljebra/image.jpg")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
U, S, VT = np.linalg.svd(gray)
```

```python
print('U:', U.shape)
print('S:', S.shape)
print('VT:', VT.shape)
print('U:', U)
```

```python
print('S:', S)
```

```python
print('VT:', VT)
```

```python
cv2_imshow(gray)
```

```python
print(np.diag(S[0:3]))
```

```python
reconstructed_img = np.matmul(np.matmul(U[:, :2], np.diag(S[:2])), VT[:2, :])
```

```python
plt.imshow(reconstructed_img, cmap='gray')
plt.show()
```

```python
for i in [5, 10, 15, 20, 30, 50]:
    reconstructed_image = np.matmul(np.matmul(U[:, :i], np.diag(S[:i])), VT[:i, :])
    plt.imshow(reconstructed_image, cmap='gray')
    plt.title("n={}".format(i))
    plt.show()
```
